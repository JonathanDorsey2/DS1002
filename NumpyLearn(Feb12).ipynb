```python
a_list = [1,2,3,4,5]
b = np.array(a_list)    # converts everything in list to strings
print(b)
type(b)                 # the type is an ndarray, not array
```

```python
print(b.shape)      # tells you the length and the size of an array
twod = [1,2,3,4,5],[3,4,5,6]
bum = np.array(twod)
print(bum.shape)
print(bum.ndim)
print(bum[0])
print(bum[0,1])
```

```python
z = [1,4,6,8,3,6,2]
z = np.array(z)
print(z > 3)
z > 3
```

```python
z[z<3]      #does the same as the above example but only includes the indeces that follow the interior logic
```

```python
first = np.array([1,2,3,4,5,6])
second = np.array([2,3,4,5,6,76])     # (Have to be numpy arrays to work! Not Python 3!)
first + second                      # This adds two arrays together
first * second                      # This multiplies two arrays together
```

```python
import numpy as np
from skimage import io as ios       # You don't need the 'as ios', Im just practicing

photo = ios.imread("https://c02.purpledshub.com/uploads/sites/62/2023/01/Giant-panda-c2c51b8.jpg?w=1029&webp=1")
type(photo)
photo.shape     # The output represents the colors: (Red, Green, Blue)
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
plt.imshow(photo[::, ::])   # (x,y, flip or no flip)
```
